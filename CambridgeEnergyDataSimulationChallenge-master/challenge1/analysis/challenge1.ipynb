## 2.Data analysis.

When using VAR model, it is very important to check whether a data is stationary or not.
Only stationary data can be used in VAR model.

I will use Hodrick–Prescott filter to check whether the data have trend components or not.
By using Hodrick–Prescott filter, we can separate(remove) a trend from a data.

note:
With regards to Hodrick-Prescott smoothing parameter, I will use 129600 as it is suggested to use 129600 for monthly data.
(A value of 1600 is suggested for quarterly data. Ravn and Uhlig suggest using a value of 6.25 for annual data and 129600 for monthly data.)
https://en.wikipedia.org/wiki/Hodrick%E2%80%93Prescott_filter
http://statsmodels.sourceforge.net/devel/generated/statsmodels.tsa.filters.hpfilter.html

```python
rang=pd.date_range('7/1/2011', '6/1/2013', freq='M')
fig, axes = plt.subplots(figsize=(18,4), ncols=3)
for i in data_unq:
    data_EN = data_uniqu[data_uniqu.House==i][['EnergyProduction']].set_index(rang)
    data_EN_smoothed, data_EN_trend = sm.tsa.filters.hpfilter(data_EN, lamb=129600)
    axes[0].plot(data_EN)
    axes[0].set_title('raw EnergyProduction')
    axes[0].set_xlabel('time')
    axes[0].set_ylabel('EnergyProduction')
    axes[1].plot(data_EN_trend)
    axes[1].set_title('trend EnergyProduction')
    axes[1].set_xlabel('time')
    axes[1].set_ylabel('EnergyProduction')
    axes[2].plot(data_EN_smoothed)
    axes[2].set_title('smoothed EnergyProduction')
    axes[2].set_xlabel('time')
    axes[2].set_ylabel('EnergyProduction')


fig, axes = plt.subplots(figsize=(18,4), ncols=3)   
for j in data_unq:
    data_DL = data_uniqu[data_uniqu.House==j][['Daylight']].set_index(rang)
    data_DL_smoothed, data_DL_trend = sm.tsa.filters.hpfilter(data_DL, lamb=129600)
    axes[0].plot(data_DL)
    axes[0].set_title('raw Daylight')
    axes[0].set_xlabel('time')
    axes[0].set_ylabel('Daylight')
    axes[1].plot(data_DL_trend)
    axes[1].set_title('trend Daylight')
    axes[1].set_xlabel('time')
    axes[1].set_ylabel('Daylight')
    axes[2].plot(data_DL_smoothed)
    axes[2].set_title('smoothed Daylight')
    axes[2].set_xlabel('time')
    axes[2].set_ylabel('Daylight')

fig, axes = plt.subplots(figsize=(18,4), ncols=3)   
for k in data_unq:
    data_TP = data_uniqu[data_uniqu.House==k][['Temperature']].set_index(rang)
    data_TP_smoothed, data_TP_trend = sm.tsa.filters.hpfilter(data_TP, lamb=129600)
    axes[0].plot(data_TP)
    axes[0].set_title('raw Temperature')
    axes[0].set_xlabel('time')
    axes[0].set_ylabel('Temperature')
    axes[1].plot(data_TP_trend)
    axes[1].set_title('trend Temperature')
    axes[1].set_ylabel('Temperature')
    axes[1].set_xlabel('time')
    axes[2].plot(data_TP_smoothed)
    axes[2].set_title('smoothed Temperature')
    axes[2].set_ylabel('Temperature')
    axes[2].set_xlabel('time')
```

As shown in graphs above, EnergyProduction, Daylight and Temperature obviously have trend components and are unstationary. (smoothed data is a data in which trend components are removed.)

As the above graphs also prove that each house has a similar wave shape, I will use smoothed data of "house 1" to take an unit root test (Augumented Dickey-Fuller test).

The unit root test tests whether a time series variable is stationary or non-stationary using an autoregressive model.
As a result of the unit root test, if a P-values of variables are small enough to reject, we can reject a null hypothesis that the precess is a unit root process and can adopt alternative hypothesis that the process is stationary.

In the unit root test, I will choose ‘AIC’ as a parameter of autolag, which will decide the number of lags that minimize the information criterium.


```python
#To take an unit root test, I create a DataFrame of smoothed data for of house1.
data_EN[1] = data_uniqu[data_uniqu.House==1][['EnergyProduction']].set_index(rang)
data_EN_smoothed_house1, data_EN_trend_house1 = sm.tsa.filters.hpfilter(data_EN[1], lamb=129600)
data_DL[1] = data_uniqu[data_uniqu.House==1][['Daylight']].set_index(rang)
data_DL_smoothed_house1, data_DL_trend_house1 = sm.tsa.filters.hpfilter(data_DL[1], lamb=129600)
data_TP[1] = data_uniqu[data_uniqu.House==1][['Temperature']].set_index(rang)
data_TP_smoothed_house1, data_TP_trend_house1 = sm.tsa.filters.hpfilter(data_TP[1], lamb=129600)
d_smoothed_house1 = {'EnergyProduction_smoothed_house1': pd.Series(data_EN_smoothed_house1), 'Daylight_smoothed_house1': pd.Series(data_DL_smoothed_house1), 'Temperature_smoothed_house1': pd.Series(data_TP_smoothed_house1)}
df_smoothed_house1 = pd.DataFrame(d_smoothed_house1)
df_smoothed_house1.head(5)
```

```python
print 'P-values of root unit test: smoothed_house1 (Hodrick-Prescott smoothing parameter = 129600)'
print 'EnergyProduction: ' + str(round(st.adfuller(df_smoothed_house1.EnergyProduction_smoothed_house1.dropna(), autolag='AIC')[1],10))
print 'Daylight:         ' + str(round(st.adfuller(df_smoothed_house1.Daylight_smoothed_house1.dropna(), autolag='AIC')[1],10))
print 'Temperature:      ' + str(round(st.adfuller(df_smoothed_house1.Temperature_smoothed_house1.dropna(), autolag='AIC')[1],10)) 
```

P-values of this root unit test cannot be said small enough to reject. 
Therefore, we cannot reject the null hypothesis, meaning that this process cannot be said stationary.

I will use another approach to remove trend components from the data, so that I can reject the null hypothesis.

There is also a way to take a first difference. 
Therefore, I will also plot and check the first differences of the three variables. And again, I will take the unit root test.(Again, I will use "house 1"  as a representative.)

```python
house1 = data_train[data_train.House==1][['EnergyProduction','Daylight','Temperature']].set_index(rang)
print 'P-values of unit root test: first  differences '
print 'EnergyProduction: ' + str(round(st.adfuller(house1.EnergyProduction.diff().dropna(), autolag='AIC')[1],100))
print 'Daylight:         ' + str(round(st.adfuller(house1.Daylight.diff().dropna(), autolag='AIC')[1],10))
print 'Temperature:      ' + str(round(st.adfuller(house1.Temperature.diff().dropna(), autolag='AIC')[1],10)) 
fig, axes = plt.subplots(ncols=3)
for i, column in enumerate(house1.columns):
    house1[column].diff().dropna().plot(ax=axes[i], figsize=(14,3), title=column+'(1-Diff)')
```

For comparison and curious, I will also plot and check the first natural logarithm differences of the three variables.   After that, I will take the unit root test again.

```python

print 'P-values of the root unit test: first natural logarithm differences '
print 'EnergyProduction: ' + str(round(st.adfuller(np.log(house1.EnergyProduction).diff().dropna(), autolag='AIC')[1],10))
print 'Daylight:         ' + str(round(st.adfuller(np.log(house1.Daylight).diff().dropna(), autolag='AIC')[1],10))
print 'Temperature:      ' + str(round(st.adfuller(np.log(house1.Temperature).diff().dropna(), autolag='AIC')[1],10)) 

fig, axes = plt.subplots(ncols=3)
for i, column in enumerate(house1.columns):
    np.log(house1[column]).diff().dropna().plot(ax=axes[i], figsize=(14,3), title=column+'(1-log-Diff)')
```

Comparing these three results of the unit root test, it is proved that taking the first difference minimizes P-values of that unit root test and the P-values are small enough to reject.
Therefore, by taking the first differences, we can reject the null hypothesis, meaning that this process can be treated as a stationary data.
Finally, the trend components is successfully removed from three variables.

To actually do the estimation, it is very important to decide a lag order.
I will check AIC, BIC and FPE to decide the lag order.

note:
The lag order is how many terms the VAR process back down.

```python
data_unique = data_uniqu[data_uniqu.House==156][['EnergyProduction','Daylight','Temperature']].set_index(rang)
data_unique_diff = data_unique.diff().dropna()
model_unique = var.VAR(data_unique_diff)
model_unique.select_order(5)
```

As shown in above table, when the lag length is 5, it minimizes AIC (and also BIC and FPE) for the VAR model.
For the cointegration, the lag length is the lag length chosen from previous criteria minus one, as I am running the model in first difference now.
Therefore, I choose 4 as the lag order.
I will plott residuals of three variables. (just for curious.)

```python
lag_order = 4
results_unique = model_unique.fit(lag_order)
results_unique.plot()
```

## 3.Prediction and evaluation.

Finaly I will predict EnergyProduction on June 2013 for each house.
First, I will use data_uniqu to make a prediction and calculate MAPE.
Next, for comparison, I will use data_train to predict EnergyProduction.

```python
def VAR_model(data_type, pred_file, mape_file):
    file_pred = open(pred_file, 'w')
    file_mape = open(mape_file, 'w')
    pred_writer = csv.writer(file_pred)
    pred_writer.writerow(['House','EnergyProduction'])
    sum0 = 0
    for i in data_type:
        data = data_train[data_train.House==i][['EnergyProduction','Daylight','Temperature']].set_index(rang)
        data_diff = data.diff().dropna()
        model = var.VAR(data_diff)
        results = model.fit(lag_order)
        data_pred = results.forecast(data_diff.values[-lag_order:],1)[0,0]+data.EnergyProduction[-1]
        sum0 += abs(data_pred-data_test.EnergyProduction[i-1])/data_test.EnergyProduction[i-1]     
        pred_writer.writerow([i,data_pred])
    mape = round(100*sum0/len(data_type),3)
    file_mape.write(str(mape))
    file_pred.close()
    file_mape.close()

    return mape
    
    
print 'MAPE: ' + str(VAR_model(data_unq, 'pred_unique.csv','mape_unique.txt')) + '% (unique houses) '
print 'MAPE: ' + str(VAR_model(range(1,501), 'pred_all.csv','mape_all.txt')) + '% (all houses)'
```
